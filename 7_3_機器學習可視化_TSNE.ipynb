```python
# K-MEANs 非監督學習 cluster
from sklearn.manifold import TSNE

#跟距離有關的 要scaling
from sklearn.preprocessing import MinMaxScaler

from sklearn.datasets import load_iris

data=load_iris()
scaler= MinMaxScaler()
data_scaled=scaler.fit_transform(data["data"])
tsne= TSNE(n_components=2, n_jobs=-1) # 降到二維 運算核心全上
data_tsne=tsne.fit_transform(data_scaled) # fit 並轉換
data_tsne
print(data_tsne.shape) #降到2維
print(data_scaled.shape)# 原本為4維資料
```

```python
#tsne圖 讓數據分離的狀況可視化 並判斷此筆資料可不可以分析
import matplotlib.pyplot as plt
import seaborn as sns
xs=data_tsne[:,0] # 全部列的第一位
ys=data_tsne[:,1] # 全部列的第二位

sns.scatterplot(x=xs,
        y=ys,
        hue=data["target"]
        )
```

```python
import pandas

df = pandas.DataFrame(dict(Feature_1=xs, Feature_2=ys, label=data["target"]))

df.plot(x="Feature_1", y="Feature_2", kind='scatter', c='label', colormap='viridis')
```

```python
# 葡萄酒資料集
from sklearn.datasets import load_wine
data_wine= load_wine()
data_wine['feature_names']
```

```python
scaler2= MinMaxScaler()
data_scaled2=scaler2.fit_transform(data_wine["data"])
tsne2= TSNE(n_components=2, n_jobs=-1)
tsne2=tsne2.fit_transform(data_scaled2)
tsne2
```

```python
xs= tsne2[:,0]
ys= tsne2[:,1]
sns.scatterplot(x=xs,
        y=ys,
        hue=data_wine["target"]
        )
```

```python
#單純降維程式碼
from sklearn import manifold, datasets
digits = datasets.load_digits(n_class=6)
X, y = digits.data, digits.target
X_tsne = manifold.TSNE(n_components=2, init='random', random_state=5, verbose=1).fit_transform(X) #verbose可以看計算過程 1 or 0
```

```python
#利用pandas快速出圖
import pandas
# dict() 將資料轉成字典形式
df = pandas.DataFrame(dict(Feature_1=X_tsne[:,0], Feature_2=X_tsne[:,1], label=y))

df.plot(x="Feature_1", y="Feature_2", kind='scatter', c='label', colormap='viridis')
```

```python
from sklearn.manifold import TSNE
import seaborn as sns
#跟距離有關的 要scaling
from sklearn.preprocessing import MinMaxScaler

from sklearn.datasets import load_breast_cancer

data= load_breast_cancer()
scaler= MinMaxScaler()
data_scaled=scaler.fit_transform(data["data"])
tsne= TSNE(n_components=2, n_jobs=-1)
data_tsne=tsne.fit_transform(data_scaled)

xs=data_tsne[:,0]
ys=data_tsne[:,1]
sns.scatterplot(x=xs,
        y=ys,
        hue=data["target"]
        )
```
