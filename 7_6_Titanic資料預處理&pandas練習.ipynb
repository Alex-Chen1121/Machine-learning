```python
#demo
'Braund, Mr. Owen Harris'.split(" ")[1].split(".")[0]

#名字處理函式
def name(s):
  s=s.split(" ")[1].split(".")[0]
  return s
```

```python
data_concated['Name']=data_concated['Name'].apply(name)
data_concated
```

```python
count = data_concated["Ticket"].value_counts()
def ticket(t):
    if pd.isna(t):
        return t
    else:
        return count[t]
data_concated["Ticket"] = data_concated["Ticket"].apply(ticket)
data_concated
```

```python
def cabin(c):
    if pd.isna(c):
        return c
    else:
        return c[0]
data_concated["Cabin"] = data_concated["Cabin"].apply(cabin)
data_concated
```

```python
c=data_concated["Name"].value_counts()
reserve= c[c>50].index

def name2(n):
  if n in reserve:
    return n
  else:
    return None

data_concated['Name']=data_concated['Name'].apply(name2)
data_concated.isna().sum()
```

```python
#one hot encoding
data_concated=pd.get_dummies(data_concated)
data_concated=pd.get_dummies(data_concated, columns=['Pclass'])
data_concated
```

```python
data_concated["Family"] = data_concated["SibSp"] + data_concated["Parch"]
data_concated
```

```python
#引用隨機森林套件
#不同的兒子餵不同的資料
#最後交給老爸投票

from sklearn.ensemble import RandomForestClassifier

clf= RandomForestClassifier()
params={
    "max_depth":range(3,11),
    "n_estimators":range(21,150,2)
}

cv= GridSearchCV(clf, params, n_jobs=-1, cv=10)
cv.fit(x_np, y)
print(cv.best_score_)
print(cv.best_params_)
```
