```python
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()
df=pd.DataFrame(iris["data"], columns= iris["feature_names"])
df
```

```python
#將數列矩陣化
import numpy as np

x=np.array(df)
print(x.shape)
```

```python
#將data歸一化(弄到0到1之間)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_scaled=scaler.fit_transform(x)
pd.DataFrame(x_scaled)
```

```python
from sklearn.cluster import KMeans

#利用k-means分三群
cluster= KMeans(n_clusters=3)
cluster.fit(x)
```

```python
#列出3群的中心點
cluster.cluster_centers_
```

```python
#列出被分到哪個群 並標記上123
cluster.labels_
```

```python
iris["target"]
```

```python
#畫出資料分布圖 用看的就大概知道可以分成幾群
import matplotlib.pyplot as plt
import seaborn as sns


sns.scatterplot(data=df,
        x="sepal length (cm)",
        y="petal length (cm)",
        hue=iris["target"])
```

```python
cluster = KMeans(n_clusters=3)
cluster.fit(x)
sns.scatterplot(data=df,
        x="sepal width (cm)",
        y="petal width (cm)",
        hue=cluster.labels_)
```

```python
# 如果我不知道k該是多少
# silhouette_score平均輪廓係數法
# 找尋組間內差異最小 組間差異最大的 分群方法
from sklearn.metrics import silhouette_score

cluster=range(2, 16) #測是要分幾群的分數最高
sc_score=[]

for k in cluster:
    test_cluster = KMeans(n_clusters=k, n_init="auto")
    test_cluster.fit(x)
    score = silhouette_score(x, test_cluster.labels_)
    sc_score.append(score)
    print(k, score)
```

```python
#畫圖來看分群k值的變化趨勢
plt.figure(figsize=(5,5)) #創建畫布
plt.plot(cluster, sc_score)
plt.rcParams['figure.figsize'] = [3,2] #調整圖片顯示大小
plt.xlabel('number of clusters',fontsize=18)
plt.ylabel('Silhouette Coefficient Score',fontsize=18)  #样本平均轮廓系数
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
#畫長條圖
plt.figure(figsize=(5,5))
plt.rcParams['figure.figsize'] = [10,10]
plt.bar(cluster, sc_score)
```
